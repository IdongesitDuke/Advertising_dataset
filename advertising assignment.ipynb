```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data=pd.read_csv('advertising.csv',sep=',')
data
```

```python
#Distribution of TV advertising spending
plt.figure(figsize=(8,4))
sns.histplot(data["TV"],bins=5)
plt.title('TV Advertising Distribution')
plt.xlabel('TV')
plt.ylabel('Frequency')
plt.show()
```

```python
data.describe()
```

```python
data.info()
```

```python
#Distribution of Radio Advertising spending
plt.figure(figsize=(8,4))
sns.histplot(data['Radio'], bins=5,kde=True,color='black')
plt.title('Radio Advertising Distribution')
plt.xlabel('Radio')
plt.ylabel('Frequency')
plt.show()
```

```python
#Distribution of Newspaper Advertising Spending
plt.figure(figsize =(8,4))
sns.histplot(data['Newspaper'], bins=5,color='purple')
plt.title('Newspaper Advertising Distribution')
plt.xlabel('Newspaper')
plt.ylabel('Frequency')
plt.show()
```

```python
#Distribution of Sales figures
plt.figure(figsize=(8,4))
sns.histplot(data['Sales'],bins=5,color='pink')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
#Correlation between TV advertising Spending and sales
sns.boxplot(data=data[['TV']])
plt.title('TV')
plt.show()
```
