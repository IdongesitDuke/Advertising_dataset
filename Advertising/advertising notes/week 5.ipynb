```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('advertising.csv', sep=',')
```

```python
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
#Distribution of Radio advertising spending
plt.figure(figsize=(10,5))
sns.histplot(data['Radio'], bins=10,color='red')
plt.title('Radio Advertising Spending Distribution')
plt.xlabel('Radio Advertising Spending')
plt.ylabel('Frequency')
plt.show()
```

```python
#Distribution of Newspaper adbertising spending
plt.figure(figsize=(10,5))
sns.histplot(data['Newspaper'],bins=10)
plt.title('Newspaper Advertising Spending Distribution')
plt.xlabel('Newspaper Advertising Spending')
plt.ylabel('Frequency')
plt.show()
```

```python
#Distribution of Sales
plt.figure(figsize=(10,5))
sns.histplot(data['Sales'],bins=10,color='brown')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
#Histograms to visualize Distibutions
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.hist(data["TV"],bins=10,color='purple',alpha=1)
plt.title('TV Spending Distribution')
plt.xlabel('TV Spending')
plt.subplot(122)
plt.hist(data['Radio'],bins=10,color='red',alpha=0.8)
plt.title('Radio Spending Distribution')
plt.xlabel('Radio Spending')

```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.hist(data['Newspaper'],bins=10, color='green',alpha=0.8)
plt.title('Newspaper Spending Distribution')
plt.xlabel('Newspaper Spending')
plt.subplot(122)
plt.hist(data['Sales'], bins=10,color='yellow',alpha=0.7)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.tight_layout()
plt.show()
```

# USING AX(AXIS) TO SET THE SUBPLOTS

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#Histograms to visualize distributions
ax[0,0].hist(data['TV'],bins=10,color='brown', alpha=0.9)
ax[0,0].set_title('TV Spending Distribution')
plt.xlabel('TV Spending')
#plt.subplot(122)
ax[0,1].hist(data['Radio'],bins=10,color='purple',alpha=0.9)
ax[0,1].set_title('Radio Spending Distribution')
plt.xlabel('Radio Spending')
#plt.figure(figsize=(10,5))
#plt.subplot(121)
ax[1,0].hist(data['Newspaper'],bins=10,color='blue',alpha=0.9)
ax[1,0].set_title('Newspaper Spending Distribution')
plt.xlabel('Newspaper Spending')
#plt.subplot(122)
ax[1,1].hist(data['Sales'],bins=10,color='red', alpha=0.9)
ax[1,1].set_title('Sales Distribution')
plt.xlabel('Sales')
plt.tight_layout()
```

```python
#Scatter plot to visualize relationships
plt.figure(figsize=(10,5))
plt.subplot(131)
plt.scatter(data['TV'],data['Sales'],color='blue',alpha=0.9)
plt.title('TV vs Sales')
plt.xlabel('TV Spending')
plt.ylabel('Sales')
plt.subplot(132)
plt.scatter(data['Radio'],data['Sales'],color='green',alpha=0.9)
plt.title('Radio vs Sales')
plt.xlabel('Radio Spending')
plt.ylabel('Radio')
plt.subplot(133)
plt.scatter(data['Newspaper'],data['Sales'],color='orange',alpha=0.9)
plt.title('Newspaper vs Sales')
plt.xlabel('Newspaper Spending')
plt.ylabel('Newspaper')
plt.tight_layout()
plt.show()
```

# Multivariate 

```python
#Correlation matrix
corr_matrix=data.corr()

#Heatmap to visualize correlations
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix,annot=True,cmap='viridis', linewidth=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Multivariate plot
#Pairplot to visualize relationships between all variables
sns.pairplot(data)
plt.show()
```

```python
#3D scatter plot of TV, Radio and Sales
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(data['TV'],data['Radio'],data['Sales'],c='r', marker='o')
ax.set_xlabel('TV')
ax.set_ylabel('Radio')
ax.set_zlabel('Sales')
plt.title('3D Scatter Plot of TV, Radio and Sales')
plt.show()
```
